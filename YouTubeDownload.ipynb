```python
#Download Youtube Videos
!pip install pytube

from pytube import YouTube
from sys import argv

link = argv[1]
yt = YouTube('https://www.youtube.com/watch?v=WXR-bCF5dbM')

print("Title: ", yt.title)

print("View: ", yt.views)

yd = yt.streams.get_highest_resolution()

# ADD FOLDER HERE
yd.download('./YTfolder')
```

```python
#Download mp3 from Youtube Video
from pytube import YouTube

yt = YouTube('https://www.youtube.com/watch?v=WXR-bCF5dbM')

yt.streams.get_audio_only().download(output_path='./YTfolder',filename=yt.title + '.mp3')
```

```python
#Download mp3 from xlsx list
from google.colab import drive
drive.mount('/content/drive')

!pip install pytube

import pandas as pd
from pytube import YouTube

data = pd.read_excel(r'/content/drive/MyDrive/MBA/DataBases/Youtube.xlsx') 
df = pd.DataFrame(data)
print(df)

df = df.reset_index()  # make sure indexes pair with number of rows

for index, row in df.iterrows():
    yt = YouTube(row['Link'])
    yt.streams.get_audio_only().download(output_path='/content/drive/MyDrive/MBA/Musicas',filename=yt.title + '.mp3')
    #print(row['Nome'], row['Link'])
```
